## 8 Queens
Consider the 8-queens problem, where the goal is to place eight queens on a chessboard such that no queen attacks any other. We encode states of this problem using a tuple of 8 digits, where the i-th digit represents the row index of a queen in column i of the chessboard. For example, `(7, 2, 6, 3, 1, 4, 0, 5)` represents the following state of the problem. Note that indices of rows and columns start from 0.      
8-queens_state.png

The following `count_attacks` function computes the number of attacks between each pair of queens in any given state. For any valid state, it will return a value between 0 and 28 (when all pairs of queens attack each other). You can use this function to define the cost/objective/fitness function. 


### Tasks

    - Hill climbing.
    - Random-restart hill climbing.
    - Simulated annealing.
    - Local beam.
    - Genetic algorithm.
- Compare the solutions of the above local search algorithms in terms of completeness, optimality, and efficiency. 

```python
def count_attacks(state):
    """
    ctr = The total number of attacks (a value between 0 and 28).
    For each pair of 2 queens:
       i1 = The row index of queen 1.
       i2 = The row index of queen 2.
       j1 = The column index of queen 1.
       j2 = The column index of queen 2.
    """
    ctr = 0
    for j1 in range(len(state)):
        for j2 in range(j1 + 1, len(state)):
            i1 = state[j1]
            i2 = state[j2]
            if i1 == i2 or j1 == j2 or (abs(i1 - i2) == abs(j1 - j2)):
                ctr += 1
    return ctr


state = (7, 2, 6, 3, 1, 4, 0, 5)
count_attacks(state)
```
